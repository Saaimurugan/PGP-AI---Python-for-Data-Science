```python
def preprocess(sent):
    wordList = []
    sent = nltk.pos_tag(nltk.word_tokenize(sent))
    #construct rule
    for word in sent:
        #print(word[0])
        if word[0] in dictionaryWords.keys():
            wordList.append(dictionaryWords.get(word[0]))
        else:
            if re.match('^[0-9]+.?[0-9]*$', word[0]):
                wordList.append('<NUM>')
            else:
                wordList.append(word[0])
    return TreebankWordDetokenizer().detokenize('<' + word + '/>' for word in wordList)
```

```python
def preprocessCPOS(sent):
    wordList = []
    sent = nltk.pos_tag(nltk.word_tokenize(sent))
    #construct rule
    for word in sent:
        #print(word[0])
        if word[0] in dictionaryWords.keys():
            wordList.append(dictionaryWords.get(word[0]))
        else:
            if re.match('^[0-9]+.?[0-9]*$', word[0]):
                wordList.append('DeantaNum>')
            else:
                wordList.append(word[0])
    return TreebankWordDetokenizer().detokenize(wordList)
```
