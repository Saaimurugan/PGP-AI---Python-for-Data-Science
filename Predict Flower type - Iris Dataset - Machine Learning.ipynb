```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.03,  random_state=100)
```

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
```

```python
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth = 3, min_samples_leaf=3)
```

```python
clf_entropy.fit(X_train, y_train)
```

```python
import pydotplus
dot_data = StringIO()
```

```python
export_graphviz(clf_entropy, out_file=dot_data, filled=True, rounded=True, special_characters=True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
```
